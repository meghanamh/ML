```python
import numpy as numpy
import matplotlib.pyplot as pyplt
x= numpy.array([1,2,3])
y=numpy.array([1.2,1.9,3.2])
sumofx=x.sum()
sumofy=y.sum()
X=(x*x).sum()
Y=(y*y).sum()
size=len(x)
w0=(X*sumofy-sumofx*Y)/(X*size-sumofx*sumofx) 
w1=(sumofy-(size*w0))/sumofx 
values=[w1*i+ w0 for i in x]
pyplt.plot(x,y,'r+')
pyplt.plot(x, values, color='green')
```
